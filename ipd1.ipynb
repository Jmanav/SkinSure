```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("pacificrm/skindiseasedataset")

print("Path to dataset files:", path)
```

```python
import os
print(os.listdir(path))
```

```python
skin_path = os.path.join(path, "SkinDisease")
print("level 1: ",os.listdir(skin_path))

skin_path1 = os.path.join(skin_path, "SkinDisease")
print("level 2: ",os.listdir(skin_path1))
```

```python
train_dir = os.path.join(skin_path1, "train")
print(os.listdir(train_dir))
test_dir = os.path.join(skin_path1, "test")
print(os.listdir(test_dir))
```

```python
#now we'll starting loading images into arrays
import cv2
import numpy as np

def load_images_from_folder(train_dir):
    X,y = [],[]
    class_names = sorted(os.listdir(train_dir)) #fixed order
    for label,disease in enumerate(class_names):
        disease_path = os.path.join(train_dir, disease)
        for file in os.listdir(disease_path):
          img_path = os.path.join(disease_path, file) #taking out the image from each folder diseases
          img = cv2.imread(img_path) #reading that image into numpy array
          if img is not None:
            img = cv2.resize(img,(64,64)) #resizes images to 64x64 pixels
            X.append(img.flatten())
            y.append(label)
    return np.array(X),np.array(y) , class_names

X_train,y_train,class_names = load_images_from_folder(train_dir)
X_test,y_test,_ = load_images_from_folder(test_dir)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

model = RandomForestClassifier(n_estimators=150, random_state=42) #estimators=number of trees, random state?
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#now lets try decision trees
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model = DecisionTreeClassifier(max_depth=20,random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy of decison trees:" , accuracy_score(y_test, y_pred))
```
